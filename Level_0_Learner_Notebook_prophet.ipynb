## Data Preperation

```python
prophet_df = df.rename(columns={"Date":"ds", "Close": "y"})
```

```python
prophet_df  
```

## Modelling

```python
m = Prophet()
m.fit(prophet_df)
```

##Forecasting

```python
future = m.make_future_dataframe(periods = 30)
forecast = m.predict(future)
```

```python
forecast
```

```python
figure = m.plot(forecast, xlabel="Date", ylabel="Price")
```
